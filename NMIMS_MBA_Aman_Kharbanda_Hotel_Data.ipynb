### 3.1.3 Who gave more special requests? Customers who cancelled the booking or customers who didnot cancel the booking? Find the percentage for both the cases (2 Marks)

```python
# Write your code here
# here we created a a new dataframe by the name 'cancelled' with the column 'is_canceled' which has the value 1, i.e they 
# cancelled the booking, also we created a dataframe 'can_req' which gave us the sum of 'total_of special_requests' by the 
#customers who cancelled the booking
cancelled = hotel[hotel['is_canceled'] == 1]
cancelled.head()
can_req =  cancelled['total_of_special_requests'].sum()
print(can_req)



```

```python
# here we created a a new dataframe by the name 'non_cancelled' with the column 'is_canceled' which has the value not equal to 1,
#i.e they did not cancel the booking, also we created a dataframe 'nonc_req' which gave us the sum of 'total_of special_requests'
# by the customers who did not cancel the booking.




non_cancelled = hotel[hotel['is_canceled'] != 1]
non_cancelled.head()
nonc_req = non_cancelled['total_of_special_requests'].sum()
print(nonc_req)
```

```python
# Calculate the percentage for both the above data by dividing by total number of special requests.
sp_req_cancel = round((can_req)/(can_req + nonc_req)*100,2)
sp_req_ncancel = round((nonc_req)/(can_req + nonc_req)*100,2)
print('percentage of customers who had cancelled the booking and had a special request -', sp_req_cancel, '%')  
print('percentage of customers who did not cancel the booking and had a special request -', sp_req_ncancel, '%' )
```

```python
# from the above two functions we can see that 21.35% customers who had cancelled the booking had special requests,where as
# 78.65% customers who did not cancel the booking had special requests, we can conclude that customers who did not cancel the
#booking had more special request.
if sp_req_ncancel > sp_req_cancel:
    print('customers who did not cancel the booking had more special request')
else:
    print('customers who cancelled the booking had more special request')
```

### 3.2.2 Create a new column "total_stay_duration" in the data frame not_cancel. (2 Marks)

Total_Stay_Duration is defined as the number of nights a customer stayed in the hotel including both weekends and weekdays

```python
# we created a new column 'total_stay_duration' and gave it a value of the sum of two columns 'stays_in_weekend_nights' &
#'stays_in_week_nights'

not_cancel['total_stay_duration'] = not_cancel['stays_in_weekend_nights'] + not_cancel['stays_in_week_nights']
```

### 3.2.3 Create a new column "total_revenue" in the data frame not_cancel. (2 Marks)

Total_Revenue can be calculated by multiplying average rental revenue(adr) with total stay duration(column created in last question)

The average daily rate (adr) measures the average rental revenue earned for an occupied room per day. The operating performance of a hotel or other lodging business can be determined by using the ADR.

```python
# we created a new column 'Total_revenue' and gave it a value of the multiplication of two columns 'adr' & 'total_stay_duration'
#'stays_in_week_nights'
not_cancel['Total_revenue'] = not_cancel['adr'] * not_cancel['total_stay_duration']
```
